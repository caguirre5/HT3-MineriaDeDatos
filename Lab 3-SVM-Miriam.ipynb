**Limpieza de datos**

Conversión de variables cuantitativas de float a int, debido a que no tenemos valores decimales en estos datos

```python
data['free sulfur dioxide']=data['free sulfur dioxide'].astype(int)
data['total sulfur dioxide']=data['total sulfur dioxide'].astype(int)
```

**Eliminación de columna 'type'**

Se ha decidido eliminar esta columna ya que consideramos que son datos no relevantes a considerar en nuestros modelos de clasificación. 

```python
data = data.drop('type', axis=1)
data
```

**Manejo de datos categoricos**

Se decidió convertir los datos categoricos a una notación binaria para mantener un formato numerico en todo el arreglo brindado para la creacion de los modelos.

```python
data['quality'] = data['quality'].replace({'Legit': 0, 'Fraud': 1})
data
```

<h2>Clasificación

**División de datos de entreno y prueba**

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_entreno, X_prueba, y_entreno, y_prueba = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

**Normalización**

```python
from sklearn.preprocessing import StandardScaler
normalizador = StandardScaler()
X_entreno = normalizador.fit_transform(X_entreno)
X_prueba = normalizador.transform(X_prueba)
```

Cristian:
- Naive Bayes
- Kernel SVM
- KNN


<H2>Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clasificador = GaussianNB()
clasificador.fit(X_entreno, y_entreno)
```

**Prediccion de los datos de prueba**

```python
y_predicciones = clasificador.predict(X_prueba)
print(np.concatenate((y_predicciones.reshape(len(y_predicciones),1), y_prueba.reshape(len(y_prueba),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
mat_conf = confusion_matrix(y_prueba, y_predicciones)
print(mat_conf)
accuracy_score(y_prueba, y_predicciones)
```

<h2>KNN

**Utilización del KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
```

<b>Encontrar valor K de menor error<b>

```python
errores = []
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_entreno, y_entreno)
    pred_i = knn.predict(X_prueba)
    errores.append(np.mean(pred_i != y_prueba))
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, 50), errores, color = 'blue', linestyle = 'dashed', marker = 'o',
         markerfacecolor = 'red', markersize = 10)
plt.title('Tasa de Error vs. Valor de K')
plt.xlabel('K')
plt.ylabel('Tasa de Error')
```

**Encontramos que el valor de K con menor porcentaje de error para nuestro modelo seria de 4. Y se inputa al KNNeighbors Classifier**

```python
# valor 4 vecinos
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_entreno, y_entreno)
pred = knn.predict(X_prueba)
```

**Calculamos la matriz de confusion y el Accuracy_Score para tomar como metricas de desempeño**

```python
mat_conf = confusion_matrix(y_prueba, pred)
print(mat_conf)
accuracy_score(y_prueba, pred)
```

<h2>Kernel SVM

```python
from sklearn.svm import SVC
clasificador = SVC(kernel = 'rbf', random_state = 0)
clasificador.fit(X_entreno, y_entreno)
```

**Prediccion con los datos de prueba**

```python
y_pred = clasificador.predict(X_prueba)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_prueba.reshape(len(y_prueba),1)),1))
```

**Calculamos la matriz de confusion y el Accuracy_Score para tomar como metricas de desempeño**

```python
mat_conf = confusion_matrix(y_prueba, pred)
print(mat_conf)
accuracy_score(y_prueba, pred)
```
